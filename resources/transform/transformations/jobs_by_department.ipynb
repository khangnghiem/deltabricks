```python
from pyspark.sql import functions as F

df = spark.sql(f"SELECT * FROM dev.job_prospects.job_1900_silver_inferred")
max_timestamp = df.agg(F.max("timestamp").alias("max_ts")).collect()[0]["max_ts"]
df = df.filter((F.col("timestamp") == max_timestamp))
df.display()
```

```python
jobs_by_department_df = df.groupBy("job_department").agg(
    F.sum("quantity").alias("count")
).orderBy(F.col("count").desc())
```

```python
display(jobs_by_department_df)
```

```python
jobs_by_department_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("dev.job_prospects.fct_jobs_by_department")
```
